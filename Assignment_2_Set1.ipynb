```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
companies=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
#Pie Plot
plt.figure(figsize=(8,8))
plt.title('Name of Companies with respect to X', fontsize = 16, fontweight = 'bold', color='black')
plt.pie(x,labels=companies,autopct='%1.0f%%')
plt.show()
```

```python
#Box Plot to find outliars
sns.boxplot(x)
```
